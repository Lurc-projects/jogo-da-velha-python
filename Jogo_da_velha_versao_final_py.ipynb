```python
def main():
    """Simula um jogo da velha,onde o jogador 1 é "X" e o jogador é "O" seguindo todas as regras de jogo, além de possuir verificaçoes da interação com usuario que informa ao mesmo caso haja alguma entrada mal formulada, garantindo que todas entradas sigam determinado padrão """
    #Inicia a sequencia de jogo
    Sequenciadejogo=True

    while Sequenciadejogo==True:
        #Reinicia a partida caso o jogador nao insira o comando de parada
        matriz=[["-","-","-"],["-","-","-"],["-","-","-"]]
        Terminodejogo1=False
        Terminodejogo2=False
        Empate=False

        #Mostra a matriz inicial ao usuario
        print(matriz[0][0],matriz[0][1],matriz[0][2])
        print(matriz[1][0],matriz[1][1],matriz[1][2])
        print(matriz[2][0],matriz[2][1],matriz[2][2])

        # Loop principal da partida

        while Terminodejogo1==Terminodejogo2==Empate: # Esta forma também funciona para este caso

            # ---- JOGADOR 1 ----
            pos=input("Jogador 1 escolha sua posicao [x,y]:")
            n=0

            while n!=7:
                #n=contador, usado para conferir se todos criterios do input estao corretos, caso isso aconteca n=7
                n=0
                if len(pos)==5:
                    n+=1
                    if pos[0]=="[":
                        n+=1
                        if pos[1] in ["0","1","2"]:
                            n+=1
                            if pos[2]==",":
                                n+=1
                                if pos[3] in ["0","1","2"]:
                                    n+=1
                                    if pos[4]=="]":
                                        n+=1
                                        # Verifica se a posição está dentro dos limites da matriz antes de tentar acessá-la
                                        if 0 <= int(pos[1]) <= 2 and 0 <= int(pos[3]) <= 2:
                                            if matriz[int(pos[1])][int(pos[3])]=="-":
                                                n+=1
                if n!= 7:
                    pos=input("Posicao invalida. Jogador1 escolha sua posicao[x,y]:")
            matriz[int(pos[1])][int(pos[3])]="X"
            #Caso o input esteja correto substituir "-" na posicao escolhida por "X"

            #mostra atual estado da matriz
            print(matriz[0][0],matriz[0][1],matriz[0][2])
            print(matriz[1][0],matriz[1][1],matriz[1][2])
            print(matriz[2][0],matriz[2][1],matriz[2][2])

            #Confere se jogador 1 ganhou
            if (matriz[0][0])==(matriz[1][0])==(matriz[2][0])=="X":
                Terminodejogo1=True
            elif(matriz[0][1])==(matriz[1][1])==(matriz[2][1])=="X":
                Terminodejogo1=True
            elif(matriz[0][2])==(matriz[1][2])==(matriz[2][2])=="X": # Removida duplicata aqui
                Terminodejogo1=True
            elif(matriz[0][0])==(matriz[0][1])==(matriz[0][2])=="X":
                Terminodejogo1=True
            elif(matriz[1][0])==(matriz[1][1])==(matriz[1][2])=="X":
                Terminodejogo1=True
            elif(matriz[2][0])==(matriz[2][1])==(matriz[2][2])=="X":
                Terminodejogo1=True
            elif(matriz[0][0])==(matriz[1][1])==(matriz[2][2])=="X":
                Terminodejogo1=True
            elif(matriz[0][2])==(matriz[1][1])==(matriz[2][0])=="X":
                Terminodejogo1=True
            # Confere empate APÓS jogada do Jogador 1 (se ele não ganhou)
            elif "-" not in matriz[0] and "-" not in matriz[1] and "-" not in matriz[2]:
                Empate=True

            # Se Jogador 1 não ganhou nem empatou, é a vez do Jogador 2
            if not Terminodejogo1 and not Empate:
                # ---- JOGADOR 2 ----
                pos=input("Jogador 2 escolha sua posicao [x,y]:")
                c=0
                #c=contador, usado para conferir se todos criterios do input estao corretos, caso isso aconteca c=7

                while c!=7:
                #Corrige possiveis erros que o usuario possa causar com um imput equivocado
                    c=0
                    if len(pos)==5:
                        c=1+c
                        if pos[0]=="[":
                            c=1+c
                            if pos[1] in ["0","1","2"]:
                                c=1+c
                                if pos[2]==",":
                                    c=1+c
                                    if pos[3] in ["0","1","2"]:
                                        c=1+c
                                        if pos[4]=="]":
                                            c=1+c
                                            # Verifica se a posição está dentro dos limites da matriz antes de tentar acessá-la
                                            if 0 <= int(pos[1]) <= 2 and 0 <= int(pos[3]) <= 2:
                                                if matriz[int(pos[1])][int(pos[3])]=="-":
                                                    c=1+c
                    if c!= 7:
                        pos=input("Posicao invalida. Jogador2 escolha sua posicao[x,y]:")
                matriz[int(pos[1])][int(pos[3])]="O"
                #Caso o input esteja correto substituir "-" na posicao escolhida por "O"

                #Mostra atual estado da matriz
                print(matriz[0][0],matriz[0][1],matriz[0][2])
                print(matriz[1][0],matriz[1][1],matriz[1][2])
                print(matriz[2][0],matriz[2][1],matriz[2][2])

                #Confere se o jogador 2 ganhou
                if (matriz[0][0])==(matriz[1][0])==(matriz[2][0])=="O":
                    Terminodejogo2=True
                elif(matriz[0][1])==(matriz[1][1])==(matriz[2][1])=="O":
                    Terminodejogo2=True
                elif(matriz[0][2])==(matriz[1][2])==(matriz[2][2])=="O": # Removida duplicata aqui
                    Terminodejogo2=True
                elif(matriz[0][0])==(matriz[0][1])==(matriz[0][2])=="O":
                    Terminodejogo2=True
                elif(matriz[1][0])==(matriz[1][1])==(matriz[1][2])=="O":
                    Terminodejogo2=True
                elif(matriz[2][0])==(matriz[2][1])==(matriz[2][2])=="O":
                    Terminodejogo2=True
                elif(matriz[0][0])==(matriz[1][1])==(matriz[2][2])=="O":
                    Terminodejogo2=True
                elif(matriz[0][2])==(matriz[1][1])==(matriz[2][0])=="O":
                    Terminodejogo2=True
                # Confere empate APÓS jogada do Jogador 2 (se ele não ganhou)
                elif "-" not in matriz[0] and "-" not in matriz[1] and "-" not in matriz[2]:
                    Empate=True

        # ---- FIM DA PARTIDA ----
        if Terminodejogo1 is True:
            #O jogador 1 ganhou, perguntar se o usuario deseja iniciar nova partidas e conferir resposta
            comando=input("Jogador 1 venceu o jogo. Deseja iniciair uma nova partida? (S/N):")
            while comando.upper() != "S" and comando.upper() !="N" : # Usar .upper() para aceitar 's' ou 'n'
                comando=input ("Comando invalido.Favor inserir novo comando (S/N):")
            if comando.upper()=="S":
                Sequenciadejogo=True # CORRIGIDO AQUI
            elif comando.upper()=="N":
                Sequenciadejogo=False

        elif Terminodejogo2 is True:
            #O jogador 2 ganhou, pergunta se o usuario deseja iniciar nova partidas e conferir resposta
            comando=input("Jogador 2 venceu o jogo. Deseja iniciair uma nova partida? (S/N):")
            while comando.upper() != "S" and comando.upper() !="N" :
                comando=input ("Comando invalido.Favor inserir novo comando (S/N):")
            if comando.upper()=="S":
                Sequenciadejogo=True
            elif comando.upper()=="N":
                Sequenciadejogo=False

        elif Empate==True:
            #O jogo empatou, perguntar se o usuario deseja iniciar nova partida e conferir resposta
            comando=input("Empate. Deseja iniciair uma nova partida? (S/N):")
            while comando.upper() != "S" and comando.upper() !="N" :
                comando=input ("Comando invalido.Favor inserir novo comando (S/N):")
            if comando.upper()=="S":
                Sequenciadejogo=True
            elif comando.upper()=="N":
                Sequenciadejogo=False

# Para rodar o jogo no Colab, você precisa chamar a função main()
main()
main()
```
